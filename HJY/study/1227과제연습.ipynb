```python
#12/27 전자계산기

#입력받기/덧셈/뺄셈/나눗셈/곱셈/괄호/제곱/루트/소수점/양수/음수/싸인/코사인/탄젠트/로그/

#입력받기 덧셈 뺄쎔 나눗셈 곱셈 제곱 괄호 종료

def calculator():
    print("간단한 전자계산기입니다")
    print("수식을 입력해주세요")
    
    while True:
    #입력받기
        expression = input("수식을 입력하세요(종료하려면 esc입력): ")
    
    #esc입력하면 종료
        if expression.lower() == 'esc':
            print("계산기를 종료합니다.")
            break

        try:
    # 수식 계산
            result = eval(expression)
            print(f"결과: {result}")
        except Exception as e:
            print(f"잘못된 입력입니다. 오류: {e}")

if __name__ == "__main__":
    calculator()
```

```python
#12/27 전자계산기

#입력받기/덧셈/뺄셈/나눗셈/곱셈/괄호/제곱/루트/소수점/양수/음수/싸인/코사인/탄젠트/로그/

#입력받기 덧셈 뺄쎔 나눗셈 곱셈 제곱 괄호 종료

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "0으로 나눌 수 없습니다."
    return a / b

while True:
    print("\n간단한 전자계산기")
    print("1: 덧셈")
    print("2: 뺄셈")
    print("3: 곱셈")
    print("4: 나눗셈")
    print("0: 종료")

    choice = input("원하는 작업을 선택하세요: ")
    if choice == '0':
        print("계산기를 종료합니다.")
        break
    elif choice in ['1', '2', '3', '4']:
        try:
            num1 = float(input("첫 번째 숫자를 입력하세요: "))
            num2 = float(input("두 번째 숫자를 입력하세요: "))
            if choice == '1':
                print("결과:", add(num1, num2))
            elif choice == '2':
                print("결과:", subtract(num1, num2))
            elif choice == '3':
                print("결과:", multiply(num1, num2))
            elif choice == '4':
                print("결과:", divide(num1, num2))
        except ValueError:
            print("유효한 숫자를 입력하세요.")
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```

```python
#12/27 전자계산기

# 입력받기/덧셈/뺄셈/나눗셈/곱셈/괄호/제곱/루트/소수점/양수/음수/싸인/코사인/탄젠트/로그/

# 입력받기 덧셈 뺄쎔 나눗셈 곱셈 
# 제곱 괄호 종료할지 더할지

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "0으로 나눌수없습니다."
    return x / y

def calculator():
    print("연산할 연산자를 선택하세요.")
    print("1. 덧셈 (+)")
    print("2. 뺄셈 (-)")
    print("3. 곱셈 (*)")
    print("4. 나눗셈 (/)")
    
    while True:
        # 사용자 입력 받기
        choice = input("원하는 연산자를 선택하세요 (1/2/3/4): ")
        
        if choice in ['1', '2', '3', '4']:
            num1 = float(input("첫 번째 숫자를 입력하세요: "))
            num2 = float(input("두 번째 숫자를 입력하세요: "))
            
            # 선택한 연산에 맞게 함수 호출
            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")

        else:
            print("잘못된 입력입니다. 다시 선택해주세요.")
        
        # 연속 계산 여부 묻기
        next_calculation = input("계속 계산하시겠습니까? (y/n): ")
        if next_calculation.lower() != 'y':
            print("계산기를 종료합니다.")
            break

# 계산기 실행
calculator()
```
